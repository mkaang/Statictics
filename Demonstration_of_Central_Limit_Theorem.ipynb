```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
sums = []

for i in range(1000):
  X = np.random.uniform(0, 1, 1000)
  sums.append(np.sum(X))
  
plt.title("CLT")
plt.hist(sums, color="purple", label = "sample")
plt.legend(loc='upper right') 
plt.show()
```
